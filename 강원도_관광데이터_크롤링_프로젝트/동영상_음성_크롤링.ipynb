```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote_plus
import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By

import openpyxl
```

```python
# 동영상용
wb = openpyxl.Workbook()
sheet = wb.active

driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.youtube.com/')

search = driver.find_element_by_name("search_query")
search.send_keys('강원도 여행')
time.sleep(1)
search.send_keys(Keys.ENTER)
time.sleep(1)

for _ in range(20):
    driver.find_element_by_tag_name('body').send_keys(Keys.END)
    time.sleep(1.7)

sheet.append(['제목', 'URL'])
    
titles = driver.find_elements_by_id('video-title')
for t in titles:
    if t.get_attribute('href'):
        title = t.text
        url = t.get_attribute('href')
        sheet.append([title, url])

wb.save('datas/videos.csv')
driver.close()
```

```python
# 음성용
wb = openpyxl.Workbook()
sheet = wb.active

driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.youtube.com/')

keywords = ['설악산', '오대산', '태백산', '치악산', '강원도 해수욕장']

sheet.append(['관광지', '제목', 'URL'])
for keyword in keywords:
    search = driver.find_element_by_name("search_query")
    search.send_keys(keyword)
    time.sleep(1)
    search.send_keys(Keys.ENTER)
    time.sleep(1)

    for _ in range(5):
        driver.find_element_by_tag_name('body').send_keys(Keys.END)
        time.sleep(1.2)

    titles = driver.find_elements_by_id('video-title')
    for t in titles:
        if t.get_attribute('href'):
            title = t.text
            url = t.get_attribute('href')
            sheet.append([keyword, title, url])
            
    search.clear()

wb.save('datas/audios.csv')
driver.close()
```
